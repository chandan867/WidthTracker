```python
nse_data['perChange']=((nse_data['CLOSE']-nse_data['LAST'])/nse_data['LAST'])*100
```

```python
nse_data.head()
```

```python
nse_data['perChange']=round(nse_data['perChange'],2)
nse_data.head()
```

```python
#sns.histplot(data=nse_data,x="perChange")
#this gives error as some %change has become infinity we need to handle this

nse_data.replace([np.inf, -np.inf], np.nan, inplace=True)
# Dropping all the rows with nan values
nse_data.dropna(inplace=True)
nse_data.head()
```

```python
plt.figure(figsize=(20,7))
sns.histplot(data=nse_data,x="perChange")
```

# a whooping nuumber of 600 stocks has 0% change
# we can remove these values considering they were not treaded to get better view of width as it distorts the graph


```python
nse_data.replace(0, np.nan, inplace=True)
# Dropping all the rows with nan values
nse_data.dropna(inplace=True)
nse_data.info()
```

```python
nse_data.head()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=nse_data,x="perChange", bins=50)
```

```python
nse_data['RS']=nse_data['perChange']+1.65
nse_data.head()
```

```python
#on 6th DEC 2021 nifty 50 closed with -1.65% change,adjusting the change to get relative change wrt to nifty
plt.figure(figsize=(10,5))
sns.histplot(data=nse_data,x="RS", bins=50)
```

# while nifty closed with -1.65% most of the stocks outperformed market
# the plot seems to normally distributed around 1.7% showing most of stocks remained sideways with very low fall
# this indicates decrease in downward momentum of market as most stocks didn't fall more than the nifty which is not seen generally in full forced bear markets
